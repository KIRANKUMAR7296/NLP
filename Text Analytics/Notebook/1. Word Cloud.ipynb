### Word Cloud

A `Graphical` Display of Words in a Corpus

`Size` of the Word is based on number of `Occurences`

Visual View of most `Popular` Terms

`Prepare` Data for Word Cloud

```python
with open("../Data/Course Descriptions.txt", 'r') as file:
    data = file.read()
    
print(f'File Data : \n{data[:130]}')
```

Create `Stopword` List

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
```

`Generate` Word Cloud Data

```python
wordcloud = WordCloud(stopwords=stopwords, 
                      max_words=25, 
                      background_color='white').generate(data)
```

`Display` the Word Cloud

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off");
```

`Enhancing` the Word Cloud

```python
print(stopwords)
```

```python
stopwords.update(['looking','help','want','hand','hands','become','popular', 
                  'use','used','learn','including','value','new','start','work'])
```
